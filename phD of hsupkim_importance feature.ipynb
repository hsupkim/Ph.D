```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_validate
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import seaborn as sns

filepath = '/content/drive/My Drive/dfMD_log_MM.csv'

dfMD_log_MM = pd.read_csv(filepath)

dfMD_log_MM_input = dfMD_log_MM.iloc[:,13:].to_numpy()
dfMD_log_MM_target = dfMD_log_MM['class2'].to_numpy()


# replace 'D' with 0 and 'M' with 1
dfMD_log_MM_target = np.where(dfMD_log_MM_target == 'D', 0, 1)

# convert the ndarray to integer type
dfMD_log_MM_target = dfMD_log_MM_target.astype(int)

dfMD_header = dfMD_log_MM.columns
x_header = dfMD_header[13:]  
print(x_header)

train_input, test_input, train_target, test_target = train_test_split(dfMD_log_MM_input, dfMD_log_MM_target, random_state=42)

xgb = XGBClassifier(subsample=0.5, n_estimators=1000, max_depth=10, learning_rate=0.1)
xgb_scores = cross_validate(xgb, train_input, train_target, return_train_score=True, n_jobs=-1)
print("Train Accuracy:", np.mean(xgb_scores['train_score']))
print("Test Accuracy:", np.mean(xgb_scores['test_score']))

xgb.fit(train_input, train_target)

feat_imp = pd.Series(xgb.feature_importances_, index=x_header)
xgb_top20 = feat_imp.sort_values(ascending=False)[:20]
print(xgb_top20)

sns.barplot(x=xgb_top20, y=xgb_top20.index)
```
