```python
#분류된 이미지를 출력하는 명령어

def draw_km1(arr, ratio=1):# n은 샘플 갯수를 나타냄
    n=len(arr)                   
    rows=int(np.ceil(n/10))       # 한 줄에 10개씩 나누어 전체 행 개수를 계산
    cols=n if rows <2 else 10    # 행이 1개이면 열의 개수는 샘플의 개수이고, 그렇지 않으면 10개임
    fig, axs=plt.subplots(rows, cols, figsize=(cols*ratio, rows*ratio), squeeze=False)
    
    for i in range(rows):
        for j in range(cols):
            if i*10+j <n:        # n개 까지만 그린다.
                axs[i,j].imshow(arr[i*5+j], cmap='gray_r')
            axs[i,j].axis('off')
    plt.show()
```

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split

train_input, val_input, train_target, val_target = train_test_split(train_input, train_target, test_size=0.2, random_state=42) #라인 22번 참조

train_scaled=train_input.reshape(-1,1600, 1600)/255.0
val_scaled=val_input.reshape(-1,1600,1600)/255.0
```

```python
print(train_scaled.shape)
```

```python
print(val_scaled.shape)
```

```python
# 합성곱 신경망 만들기
#커널 크기가 3인 필터를 32개 만듦
#여기서 주의해야할 점은 input_shape가 3차원인것에 주의

cnn=keras.Sequential()
cnn.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=(1600,1600,1)))

#커널 크기가 4인 32개의 필터를 사용
```

```python
cnn.add(keras.layers.MaxPooling2D(2)) #풀링은 최대와 평균이 있는데, 전형적인(2,2) 풀링을 사용해서 이미지 크기를 절반으로 줄일 수 있음
```

```python
cnn.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same'))
cnn.add(keras.layers.MaxPooling2D(2))

#첫번째 합성곱-풀링층에 두번째 합성곱-풀링층을 추가했음
```

```python
#이진분류일때는 sigmoid, 다중분류일때는 softmax를 사용

cnn.add(keras.layers.Flatten())
cnn.add(keras.layers.Dense(32, activation='relu'))
cnn.add(keras.layers.Dropout(0.4))
cnn.add(keras.layers.Dense(2,activation='softmax'))
cnn.summary()
```

```python
keras.utils.plot_model(cnn, show_shapes=True)
```

```python
#모델 컴파일 할 때 2진분류일때는 loss값을 binary_crossentropy로 해야함
# 다중분류일 때는 sparse_categorical_crossentropy를 사용

cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
cb = keras.callbacks.ModelCheckpoint('best-model.h5', save_best_only=True)
es = keras.callbacks.EarlyStopping(patience=2, restore_best_weights=True)
history = cnn.fit(train_scaled, train_target, epochs=20, validation_data=(val_scaled, val_target), callbacks=[cb, es])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train','val'])
plt.show()
```

```python
from sklearn.cluster import KMeans

img_np_2d=img_np.reshape(-1, 400*1800) # 3차원 배열을 2차원 배열(샘플갯수, 높이x너비)로 변환

inertia =[] # 군집분석을 할 때 군집을 몇개로 구분할지 알아내기 위해서 inertia라는 빈 리스트 작성

for k in range(2,7):
    km=KMeans(n_clusters=k, random_state=42)
    km.fit(img_np_2d)
    inertia.append(km.inertia_)

    
plt.plot(range(2,7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```
